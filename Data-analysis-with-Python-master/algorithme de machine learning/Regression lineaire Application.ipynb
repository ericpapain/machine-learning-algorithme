Pour la regression linéaire, nous avons juste une seule variable prédictive qui permet de voir les valeurs a prédire

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv("algo/dataset/univariate_linear_regression_dataset.csv")
```

```python
X = df.iloc[0:len(df),0]
Y = df.iloc[0:len(df),1]
```

##a present on utilise les bibliothèques de matplotlib pour representer le nuage de point correspondant a nos données

```python
axes = plt.axes()
```

```python
axes = plt.axes()

axes.grid() 
```

```python
plt.scatter(X,Y)
```

```python
axes = plt.axes()
axes.grid()
plt.scatter(X,Y)
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)
```

```python
print (slope)
```

```python
print (intercept)
```

```python
print (r_value)
```

```python
print (std_err)
```

Après que la fonction linregress() nous ait renvoyé les paramètres de notre modèle : slope et intercept, on pourra effectuer des prédictions. En effet, la fonction de prédiction sera de la forme :

F(x) = slope * x  + intercept

```python
def predict(x):
   return slope * x + intercept
```

```python
#la variable fitLine sera un tableau de valeurs prédites depuis la tableau de variables X

fitLine = predict(X)
plt.plot(X, fitLine, c='r')
```

```python
axes = plt.axes()
axes.grid()
plt.scatter(X,Y)

fitLine = predict(X)
plt.plot(X, fitLine, c='r')
```
